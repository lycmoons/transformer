```python
import matplotlib.pyplot as plt
import numpy as np
import yaml

with open("/data/transformer/config.yaml", "r") as f:
    params = yaml.safe_load(f)

lr_histories_10 = np.load(f"{params['outputs_dir']}lr_histories_10.npy")
lr_histories_20 = np.load(f"{params['outputs_dir']}lr_histories_20.npy")
lr_histories_30 = np.load(f"{params['outputs_dir']}lr_histories_30.npy")
loss_histories_10 = np.load(f"{params['outputs_dir']}loss_histories_10.npy")
loss_histories_20 = np.load(f"{params['outputs_dir']}loss_histories_20.npy")
loss_histories_30 = np.load(f"{params['outputs_dir']}loss_histories_30.npy")

lr_histories = np.concatenate((lr_histories_10, lr_histories_20, lr_histories_30))
loss_histories = np.concatenate((loss_histories_10, loss_histories_20, loss_histories_30))
```

<br>

---

### 训练过程中学习率的调度

```python
plt.figure(figsize=(16, 4))
steps = np.arange(1, len(lr_histories) + 1)
plt.plot(steps, lr_histories)
plt.xlabel("num_steps")
plt.ylabel("learning_rate")
plt.title("Learning Rate Schedule")
plt.grid(True)
plt.show()
```

<br>

---

### 训练过程中损失值的变化

```python
plt.figure(figsize=(16, 4))
epochs = np.arange(1, len(loss_histories) + 1)
plt.plot(epochs, loss_histories)
plt.xlabel("num_epochs")
plt.ylabel("loss")
plt.title("Training Loss Curve")
plt.grid(True)
plt.show()
```
